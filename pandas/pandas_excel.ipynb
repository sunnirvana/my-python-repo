```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

## 读入Excel文件方式

### - pands.read_excel(excel_file, sheet_name, index_col)

```python
excel_file = './movies.xls'
movies = pd.read_excel(excel_file)
```

```python
movies_sheet1 = pd.read_excel(excel_file, sheet_name=0, index_col=0)
movies_sheet1.head()
```

```python
movies_sheet2 = pd.read_excel(excel_file, sheet_name=1, index_col=0)
movies_sheet2.head()
```

```python
movies_sheet3 = pd.read_excel(excel_file, sheet_name=2, index_col=0)
movies_sheet3.head()
```

```python
movies = pd.concat([movies_sheet1, movies_sheet2, movies_sheet3])
movies.shape
```

### - pandas.ExcelFile(excel_file).parse(sheet_name)

```python
xlsx = pd.ExcelFile(excel_file)
movie_sheets = []
for sheet in xlsx.sheet_names:
    movie_sheets.append(xlsx.parse(sheet))
movies = pd.concat(movie_sheets)
movies.shape
```

## 显示记录 (records)

```python
# 显示前几行 (默认5行)
movies.head()
```

```python
# 显示尾几行 (默认5行)
movies.tail()
```

```python
# 排序
sorted_by_gross = movies.sort_values(['Gross Earnings'], ascending=False)
sorted_by_gross['Gross Earnings'].head(10)
```

```python
sorted_by_gross['Gross Earnings'].head(10).plot(kind='barh')
plt.show()
```

```python
sorted_by_gross['IMDB Score'].plot(kind='hist')
plt.show()
```

```python
# 只读一部分列 
movies_subset_columns = pd.read_excel(excel_file, usecols=6) # 0 -> index:6 (前7列)
movies_subset_columns.head()
```

```python
# 新建一列 -- 净利润
movies['Net Earnings'] = movies['Gross Earnings'] - movies['Budget']
movies.head()
```

```python
# movies[['Net Earnings']] 得到新的DataFrame, 如果使用 movies['Net Earnings'] 得到是Series
sorted_movies = movies[['Net Earnings']].sort_values(['Net Earnings'], ascending=False)
sorted_movies.head(10)['Net Earnings'].plot.barh()
plt.show()
```

```python
# 可以直接使用movies做sort_values
sorted_movies = movies.sort_values(['Net Earnings'], ascending=False)
sorted_movies.head(10)['Net Earnings'].plot.barh()
plt.show()
```

## 数据透视表 (Pivot Table)

```python
movies_subset = movies[['Year', 'Gross Earnings', 'Duration']]
movies_subset.head()
```

```python
# 按时长和年份分
durations_by_year = movies_subset.pivot_table(index=['Year'])
durations_by_year[['Duration']].plot()
plt.show()
```

```python
# 不同的中心，图例不同
durations_by_year = movies_subset.pivot_table(index=['Duration'])
durations_by_year[['Year']].plot()
plt.show()
```

```python
# 使用 Year 作为index的透视表
earnings_by_year = movies_subset.pivot_table(index=['Year'])
earnings_by_year.head()
```

```python
# Plot
earnings_by_year.plot()
plt.show()
```

```python
earnings_by_year[['Gross Earnings']].plot()
plt.show()
```

```python
movies_subset = movies[['Country', 'Language', 'Gross Earnings']]
movies_subset.head()
```

```python
earnings_by_co_lang = movies_subset.pivot_table(index=['Country', 'Language'])
earnings_by_co_lang.head()
```

```python
earnings_by_co_lang.head(20).plot(kind='bar', figsize=(20,8))
plt.show()
```

## 导出Excel

```python
# 1. excel
movies.to_excel('output.xlsx', index=False)
```

```python
# 2. advanced
writer = pd.ExcelWriter('output_adv.xlsx', engine='xlsxwriter')
movies.to_excel(writer, index=False, sheet_name='report')
workbook = writer.book
worksheet = writer.sheets['report']

# 定制格式
header_fmt = workbook.add_format({'bold':True})
worksheet.set_row(0, None, header_fmt)

# 写入
writer.save()
```
